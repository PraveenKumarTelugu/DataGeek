```python
# Importing the library
import pandas as pd
```

```python
# Importing the train dataset
train = pd.read_csv("D:\Technology\Meetup\Training_Set3.csv")
```

```python
#Importing the label data set
labels = pd.read_csv("D:\Technology\Meetup\Training_Set_Label.csv")
```

```python
#Merge the train and labels dataset into train
train = pd.merge(train,labels)
```

```python
y = train['status_group']
```

```python
#Working with the date format
#from datetime import date
#from functools import partial
#convert_to_date = partial(date,month=1,day=1)
#for i in range(len(train)):
    #train['construction_year'][i] = convert_to_date(train['construction_year'][i])
```

```python
# Working with date difference from date_recorded to construction_year
from datetime import datetime
train['construction_year'] = [datetime.strptime(i, '%d-%m-%Y') for i in train['construction_year']]
train['date_recorded'] = [datetime.strptime(i, '%d-%m-%Y') for i in train['date_recorded']]
delta = [(train['date_recorded'][i] - train['construction_year'][i]) for i in range(len(train))]
delta = pd.DataFrame(delta)
delta = delta.astype('timedelta64[D]')
delta = delta.astype(int)
train['Numdays'] = delta
```

```python
# Encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le_statusgroup = LabelEncoder()
y = le_statusgroup.fit_transform(y)
```

```python
# Extracting the categorical features
encoding_columns = ['basin','extraction_type_class','management_group','payment_type','water_quality','quantity_group','source','waterpoint_type_group']
```

```python
encode_training_data = train[encoding_columns]
```

```python
encode_training_data
```

```python
# Encoding the categorical data
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
encoded_training_data = encode_training_data.apply(LabelEncoder().fit_transform)
hotencoding = OneHotEncoder(categorical_features = 'all')
encode_training_data = hotencoding.fit_transform(encode_training_data).toarray()
```

```python
encoded_training_data
```

```python
train[encoding_columns] = encoded_training_data[encoding_columns]
```

```python
features = ['amount_tsh','gps_height','longitude','latitude', 'basin','region_code','population','extraction_type_class','management_group','payment_type','quantity_group','waterpoint_type_group','Numdays']
train = train[features]
```

```python
train.head(5)
```

```python
train.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size = 0.30, random_state = 0)
```

```python
str(train)
```

```python
# Fitting random_forest to the training set
from sklearn.ensemble import RandomForestClassifier as rfc
clf = rfc(n_estimators = 100, random_state = 0)
clf.fit(X_train, y_train)
y_pred= clf.predict(X_test)
```
